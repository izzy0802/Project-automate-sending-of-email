# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui
import pyperclip
import time
import pandas as pd
import numpy as np

pyautogui.PAUSE = 1
#abrir o navegador
#pyautogui.press("winleft")
#pyautogui.write("Edge")
#pyautogui.press("enter")
pyautogui.alert("Vai começar, aperte OK e não mexa em nada")
pyautogui.hotkey('ctrl','t')

#abrir o drive
#ensinar aqui o write
link = "https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga"
pyperclip.copy(link)
pyautogui.hotkey('ctrl','v')
pyautogui.press("enter")
time.sleep(5)

pyautogui.click(x=539, y=380, clicks=2)
time.sleep(2)

pyautogui.click(x=513, y=436)#clicar no arquivo
pyautogui.click(x=1678, y=246)#clicar nos 3 pontos
pyautogui.click(x=1413, y=710)# clicar no fazer download
time.sleep(5)
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento

- Quantidade de Produtos

```python
tabela = pd.read_excel(r"C:\Users\isis.santos\Downloads\Vendas - Dez.xlsx")
display(tabela)

faturamento = np.sum(tabela['Valor Final'])
quantidade = np.sum(tabela['Quantidade'])
```

### Vamos agora enviar um e-mail pelo gmail

```python
#abre a pagina e abre o email
pyautogui.hotkey('ctrl','t')
pyperclip.copy("https://outlook.live.com/mail/0/")
pyautogui.hotkey('ctrl','v')
pyautogui.press("enter")
time.sleep(7)

#clica em escrever mensagem
pyautogui.click(x=288, y=227)
time.sleep(2)

#escreve a pessoa a quem será enviado o email, assunto e corpo
pyautogui.write("isistristao000@outlook.com")
pyautogui.press("tab")
pyperclip.copy("Relatório de Vendas")
pyautogui.hotkey('ctrl','v')
pyautogui.press("tab")

text = f"""
Prezados, bom dia

O faturamento de ontem foi de: R$ {faturamento:,.2f}
Aquantidade de produtos foi de: {quantidade:,}

abs
Isis Tristao"""

#envia o email
pyperclip.copy(text)
pyautogui.hotkey('ctrl','v')
pyautogui.hotkey('ctrl','enter')
```
